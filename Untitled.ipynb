```python
from datetime import timedelta, datetime

def convert_partial_year(number):

    year = int(float(number))
    d = timedelta(days=(float(number) - year)*365)
    day_one = datetime(year,1,1)
    date = d + day_one
    return date
```

```python
dfx["try"]=dfx["year+fraction"].apply(lambda x:convert_partial_year(x))
```

```python
dfy=dfx[["altimeter_type","merged_file_cycle","year+fraction","number_obsv","number_weighted_obsv","gmsl","std_dev_gmsl",
     "smoothed_gmsl","gmsl_variation","std_dev_gmsl_variation","smoothed_gmsl_variation_1","smoothed_gmsl_variation_2"]].astype(float)
```
